# **Step - 1** : Read CSV

```python
import pandas as pd
import numpy as np
data = pd.read_csv("50_Startups_MLR.csv")
data.head()
```

# **Step - 2** : Separate both i/p & o/p

## Independent Variables - R&D Spend, Administration, Marketing Spend, State

```python
X = data.iloc[:,0:4].values
```

## Dependent Variables - Profit

```python
Y = data.iloc[:, -1].values
```

# **Step - 3** : Handle Missing Data

```python
data['Marketing Spend'] = data['Marketing Spend'].replace(0,np.nan)
data['R&D Spend'] = data['R&D Spend'].replace(0,np.nan)
data['Administration'] = data['Administration'].replace(0,np.nan)
data.tail
```

```python
from sklearn.impute import SimpleImputer
values=[0,0.00]
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X[:, 0:3] = imputer.fit_transform(X[:, 0:3])
print(X)
```

# **Step - 4** : Handle Categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
composer = ColumnTransformer(transformers= [('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X =composer.fit_transform(X)
print(X)
```

# **Step - 5** : Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X[:, 0:3] = scaler.fit_transform(X[:, 0:3])
print(X)
```

# **Step - 6** : Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```
